```python
class LogisticRegression:
   def Train_data(x,y,learning_rate,no_of_iterations):
     #y 's shape = (30000,1)
     # x's shape =(30000,784)
    m=x.shape[0]#training
    n=x.shape[1]#features
    w=np.zeros((n,1))
    b= 0
    cost=[]
    for i in range(no_of_iterations):
      z=np.dot(x,w)+b#(30000,1)
      y_pred=1/(1 + np.exp(-z))
      cost_i = (-1/(m))*np.sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred),axis = 0)
      a= y_pred - y #(m,1)
      a1 = a.transpose()#(1,m)
      derivative_w=(1/m)*np.dot(a1,x)#(1,n)
      derivative_w_T=derivative_w.transpose()
      derivative_b=(1/m)*np.sum(y_pred-y,axis = 0,keepdims = True) # np.array([[1],[2],[3]])  python convert into np.array([1,2,3])
      w=w-(learning_rate)*(derivative_w_T)
      b=b-(learning_rate)*derivative_b
      cost.append(cost_i)
     
      if(i%math.ceil(no_of_iterations/100) == 0):
           print('Cost is:',cost_i,'after ',i,'iterations')
            # z_Test=np.dot(x,w)+b
            # Y_Test=1/(1 + np.exp(-z_Test))
            # plt.scatter( x[:,5], y, c='r', marker='x' )        
            # plt.plot(x[:,5], Y_Test , '-o')
            # plt.xlabel('X →')
            # plt.ylabel('Y →')
            # plt.show()
      
    return w,b,cost
   def predict_label(X_test,w,b):
     k=b.shape[0]
     m=X_test.shape[0]
     probability=np.zeros((10,m,1))
    #  wt=w.transpose()
     for i in range(k):
      #  wtk=w[k].transpose()
       z=np.dot(X_test,w[i])+b[i]
       y_pred=1/(1 + np.exp(-z))
       probability[i]=y_pred
    #  probability=np.delete(probability,0,axis=1) 
     final_results=(np.argmax(probability, axis=0))  
     return final_results


```
